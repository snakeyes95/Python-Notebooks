```python
#stemming and lemmatization in nlp
#stemming is basically reducing the word to its base or root form
#eg intelligently intelligent -> intelligen
#goes going gone -> go
#final finally finalize -> final
#etc
#the problem with this that the intermediate representation of the word or the root word may not have a meaning in many cases.
#lemmatization is the same as stemming but the only difference here is that the final intermediate representation will have some meaning
#eg the final intermediate representation instead of intelligen would be intelligent
#lemmatization will take more time then stemming 
#use lemmatization when the meaning of words are important eg ques ans system
#use stemming when the meaning of the word is not important eg spam detection.
#eg a chatbot will have to work with lemmatization
#stemming and lemmatization may lead to loss of information.
para="""I should not obtrude my affairs so much on the notice of my
readers if very particular inquiries had not been made by my
townsmen concerning my mode of life, which some would call
impertinent, though they do not appear to me at all impertinent,
but, considering the circumstances, very natural and pertinent. Some
have asked what I got to eat; if I did not feel lonesome; if I was not
afraid; and the like. Others have been curious to learn what portion
of my income I devoted to charitable purposes; and some, who have
large families, how many poor children I maintained. I will
therefore ask those of my readers who feel no particular interest in
me to pardon me if I undertake to answer some of these questions in
this book. In most books, the I, or first person, is omitted; in
this it will be retained; that, in respect to egotism, is the main
difference. We commonly do not remember that it is, after all,
always the first person that is speaking. I should not talk so much
about myself if there were anybody else whom I knew as well.
Unfortunately, I am confined to this theme by the narrowness of my
experience. Moreover, I, on my side, require of every writer, first or
last, a simple and sincere account of his own life, and not merely
what he has heard of other men's lives; some such account as he
would send to his kindred from a distant land; for if he has lived
sincerely, it must have been in a distant land to me. Perhaps these
pages are more particularly addressed to poor students. As for the
rest of my readers, they will accept such portions as apply to them. I
trust that none will stretch the seams in putting on the coat, for
it may do good service to him whom it fits."""

from nltk.stem import PorterStemmer#import the PorterStemmer class
#tokeninzing our dataset
sentences=nltk.sent_tokenize(para)
print(sentences[:1])
#creating an object of the PorterStemmer class
stemmer=PorterStemmer()
```

```python
#performing stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])#will remove any extra spaces as well
    words=[stemmer.stem(i) for i in words]
    sentences[i]=' '.join(words)
print(sentences[:5])#we can see that some of the spellings like notice was etc has been stemmed successfully
```

```python
#performing lemmatization on our data
from nltk.stem import WordNetLemmatizer
#tokenizing sentences
sentences=nltk.sent_tokenize(para)
#creating an object of the WordNetLemmatizer
lemma=WordNetLemmatizer()
print(sentences[-4:])
```

```python
#preforming lemmatization
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])#splitting each sentence into words
    words=[lemma.lemmatize(i) for i in words]
    sentences[i]=' '.join(words)
sentences[-4:]
```

```python
print(sentences[:5])#we can see unlike stemming the spelling of notice large etc has not been changed ie the intermediate 
#word representation is still meaningful
```

```python
#generating ngrams from text using nltk
str1="""If we think about it, there are numerous benefits of science and technology. They range from the little things to the big ones. For instance, the morning paper which we read that delivers us reliable information is a result of scientific progress. In addition, the electrical devices without which life is hard to imagine like a refrigerator, AC, microwave and more are a result of technological advancement.

Furthermore, if we look at the transport scenario, we notice how science and technology play a major role here as well. We can quickly reach the other part of the earth within hours, all thanks to advancing technology.

In addition, science and technology have enabled man to look further than our planet. The discovery of new planets and the establishment of satellites in space is because of the very same science and technology.
Similarly, science and technology have also made an impact on the medical and agricultural fields. The various cures being discovered for diseases have saved millions of lives through science. Moreover, technology has enhanced the production of different crops benefitting the farmers largely."""
```

```python
from nltk.tokenize import word_tokenize
from nltk.text import Text
from nltk.util import ngrams
words=word_tokenize(str1)
final=Text(words)
ngram_data=[]
for word in words:
    ngram_data.append(list(ngrams(word,8,pad_left=True,pad_right=True,left_pad_symbol='_',right_pad_symbol='_')))
ngram_data
```

```python
#flattening list since ngram data is currently a list of lists
ngram_data_updated=[temp for data in ngram_data for temp in data]
ngram_data_updated[:20]#we will get a flat list of 8 gram data
```

```python
#generating the ngram values.
temp=ngram_data_updated
for idx,val in enumerate(ngram_data_updated):
    temp[idx]=''.join(val)
print(temp)
```

```python
#sorting ngrams based on freq
freq={}
for i in temp:
    if i not in freq:
        freq.update({i:1})
    else:
        freq.update({i:freq[i]+1})
freq
```

```python
sorted(freq.items(),key=lambda x:x[1],reverse=True)#sorting our dictionary based on highest freq mostly use top 300 ngrams
```
